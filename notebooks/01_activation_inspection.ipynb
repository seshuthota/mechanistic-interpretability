# Week 1: Activation Inspection (GPT-2)

Goal: cache one residual-stream activation and record summary stats.

```python
import random
import numpy as np
import torch
from transformer_lens import HookedTransformer
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

MODEL_NAME = 'gpt2'
PROMPT = 'When John and Mary went to the store, John gave a bottle to'
HOOK_NAME = 'blocks.5.hook_resid_post'
```

```python
model = HookedTransformer.from_pretrained(MODEL_NAME)
tokens = model.to_tokens(PROMPT)
logits, cache = model.run_with_cache(tokens)
resid = cache[HOOK_NAME].detach().float().cpu()
```

```python
summary = {
    'model': MODEL_NAME,
    'prompt': PROMPT,
    'hook': HOOK_NAME,
    'shape': list(resid.shape),
    'mean': float(resid.mean().item()),
    'std': float(resid.std().item()),
    'l2': float(torch.linalg.vector_norm(resid).item()),
}
summary
```
